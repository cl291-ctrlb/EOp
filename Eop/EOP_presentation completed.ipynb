```python
#List Comprehensions (compact way to create lists using a single line.)
#Create a list of squares of numbers 1â€“5
squares = [x*x for x in range(1, 6)]
print(squares)


#what is compressions? (a short and powerful way to create lists, sets, or dictionaries using a single line of code â€” instead of writing loops. )
```

```python
#Nested List Comprehensions
#Example: Create a 3x3 matrix
matrix = [[i for i in range(1, 4)] for j in range(3)]
print(matrix)
```

```python
#Dictionary Comprehensions (Short syntax to create/update dictionaries.)
#Number and its square
squares_dict = {x: x*x for x in range(1, 6)}
print(squares_dict)
```

```python
#Functions (reusable blocks of code.)
def greet():                          # defining a function
    print("Hello Python!")
greet()                               # calling a function
```

```python
#Default Parameters (If no value is passed â†’ uses default argument.)
def welcome(name="Guest"):
    print("Welcome", name)

welcome("Nitin")
welcome()
```

```python
#Variable Arguments (Used when number of inputs may vary.)

#âƒ£ *args â†’ Multiple positional arguments (Used when we want to pass many values without specifying count)
def add_numbers(*args):   # args (collects all values into a tuple)
    total = 0        # start with 0
    for num in args:       
        total += num      #adds each number to total
    return total

print(add_numbers(10, 20))
print(add_numbers(5, 15, 25, 35))
```

```python
#kwargs â†’ Multiple keyword arguments (Stores values as a dictionary (key:value))
def student_details(**kwargs):     #kwargs allows us to pass multiple keyword arguments (collects them into a dictionary)
    for key, value in kwargs.items(): # badla lia jaega
        print(key, ":", value)

student_details(name="Nitin", age=20, course="Python")
```

```python
#Specialized Sorts
#Sorting using a custom key
#Sort list of tuples by 2nd element
students = [("Nitin", 85), ("Amit", 95), ("Ravi", 75)]  #list of tuples
students_sorted = sorted(students, key=lambda x: x[1], reverse=True)  #sorted() function sorts the list.  #reverse=True(sorts in descending order)
print(students_sorted)
```

```python
#Python Collections is a specialized library that provides alternative container data structures (more powerful than normal list, tuple, dict)
from collections import * #First import
```

```python
#namedtuple()            (a tuple, but allows accessing elements using names instead of indexes.)
from collections import namedtuple

Student = namedtuple("Student", ["name", "age"])
s = Student("Nitin", 20)
print(s.name, s.age)
```

```python
#deque (Double-ended Queue)             (Faster append and pop operations from both left and right)
from collections import deque

dq = deque([10, 20, 30])  #deque is a data structure that allows insertion and deletion of elements from both ends â€” front and rear.
dq.append(40)
dq.appendleft(5)
print(dq)
```

```python
# ChainMap             (Combines multiple dictionaries and treats them as one.)
from collections import ChainMap

dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}
cm = ChainMap(dict1, dict2)  # Search order â†’ dict1 â†’ dict2
print(cm["b"], cm["c"])
```

```python
#Counter          (Counts frequency of elements)
from collections import Counter

c = Counter("banana")
print(c)
```

```python
#OrderedDict     #OrderedDict is a special type of dictionary from the collections module that remembers the order in which the items were inserted.
from collections import OrderedDict

od = OrderedDict()  #creates an empty ordered dictionary.
od["A"] = 10          #Adding items
od["B"] = 20
print(od)
```

```python
#defaultdict      (Provides a default value for missing keys â†’ avoids KeyError.)
from collections import defaultdict

dd = defaultdict(int)
dd["a"] += 1             # += 1 (0 + 1 = 1)
print(dd["a"], dd["b"])
```

```python
#UserDict  (Allows creation of our own customized dictionary behavior.)
from collections import UserList


# Define a subclass of UserList so we can customize list behaviour
class MyList(UserList):
    def remove(self, value):         ## Override the remove method of the list interface
        print("Removal not allowed!")       #Override means to replace or change the usual behavior of something

l = MyList([1, 2, 3])
l.remove(2)
```

```python
#UserString (Custom string class for modifying string behaviors.)
from collections import UserString            #allows you to create custom list-like classes by inheriting from it.
                                              #UserList behaves like a list but can be modified or extended easily.
class MyString(UserString):   #Defining a subclass
    def upper(self):           # Override the 'upper()' method (normally converts a string to uppercase)
        return "Uppercase not allowed!"

s = MyString("python")
print(s.upper())
```

```python
#Tkinter is Pythonâ€™s standard GUI library â†’ helps create windows, buttons, text boxes, etc.
from tkinter import *                       #Import everything (*) from the tkinter module.
#Every GUI program starts with a root window.
from tkinter import *

root = Tk()  # Create the main/root window.
root.title("My First GUI") #Set the title of the window
root.geometry("300x200")  # width x height

root.mainloop() #Start the Tkinter event loop.

#Output (Window Display):
#A simple blank window titled My First GUI, size 300Ã—200.
```

```python
#Components
#Widgets (UI components (Button, Entry, Label, etc.))
#Events   (actions (button click, key press) #Actions performed by the user

import tkinter as tk              #Import the tkinter module and give it a short alias 'tk'
                                   #helps avoid writing 'tkinter.' repeatedly.


# Creating main window
root = tk.Tk()                        #Create the main application window
root.title("Widgets & Events Example")             #Set the title text on the window's title bar.
 
# Label Widget            #(for text display)
label = tk.Label(root, text="Enter your name:")          #Create a label widget with text.
label.pack()                                               #'pack()' adds it to the window (centers it vertically).

# Entry Widget (for text input)
entry = tk.Entry(root)            #Create an Entry widget (a text box where user can type).
entry.pack()             #Place it below the label.

# Event Function
def show_name():             #function runs when the user clicks the "Submit" button.
    user_name = entry.get()        #Get text entered by user from the Entry widget.
    output_label.config(text="Hello, " + user_name + "!")

# Button Widget (Triggers event on click)
button = tk.Button(root, text="Submit", command=show_name)
button.pack()           #Add the button below the Entry widget.

# Label to display output
output_label = tk.Label(root, text="")           #Create an empty label to display the result later.
output_label.pack()                                #Add it to the window.


# Main loop
root.mainloop()

# âœ… Starts the Tkinter main loop.
# âœ… Keeps the window open and waits for user events (clicks, typing, etc.).
# âœ… Without this, the window would close immediately.
```

```python
#  Example of Event Binding â€” Key Press Event
 #Event binding means linking (binding) a specific user action (event)

import tkinter as tk   #Import the tkinter module for GUI development
root = tk.Tk()
root.title("Key Event Example")

def on_key_press(event):          #function is called automatically whenever a key is pressed.
    print("You pressed:", event.char)   #'event' parameter stores information about the event, like:
                                           # which key was pressed, which widget triggered it, etc.

root.bind("<KeyPress>", on_key_press)              #"<KeyPress>" means â€” whenever ANY key is pressed while the window is active,
                                                   # call the function 'on_key_press'.

root.mainloop()
```

```python
#Adding a Label and Button
from tkinter import *           #Import all Tkinter classes and functions

 
def say_hello():        #function is called when the user clicks the button.
    label.config(text="Hello, Nitin!")         #changes the text of the label using .config()

root = Tk()
label = Label(root, text="Click the button!")         #Create a Label widget    #'root' means the Label is placed inside the main window
label.pack()                                            #'text' sets the initial message shown on the label

btn = Button(root, text="Greet", command=say_hello)           #Create a Button widget
                                                                #'command=say_hello' means that when the button is clicked,
                                                                # Tkinter will automatically call the function say_hello()
btn.pack()                                           #.pack() places the button below the label.


root.mainloop()                #This keeps the window open and listens for user events (like button clicks)
```

```python
#Text Widget (Multiline Text)
from tkinter import *

root = Tk()
text = Text(root, height=5, width=30) #'Text' widget allows multi-line text input and editing.
                                         #'Text' widget allows multi-line text input and editing.
                                           #width  = number of characters per line (approx 30)             
                                             
text.insert(END, "Write here...") #insert(index, text) adds text at a specific position.
                                #'END' means add text at the end of the current content.
text.pack()
root.mainloop()
```

```python
#INSERT Operation
import sqlite3

conn = sqlite3.connect('students.db')
cur = conn.cursor()

#we are inserting a new record with name = 'Nitin' and marks = 90
p=cur.execute("INSERT INTO student(name, marks) VALUES('Nitin', 90)")
conn.commit()  # Save changes
conn.close()

import os
db_path = os.path.abspath('students.db')
print(db_path)
```

```python
#READ (SELECT Query)
import sqlite3

conn = sqlite3.connect('students.db')
cur = conn.cursor()

cur.execute("SELECT * FROM student")
rows = cur.fetchall()

for row in rows:
    print(row)

conn.close()
```

```python
#UPDATE Operation
import sqlite3

conn = sqlite3.connect('students.db')
cur = conn.cursor()

cur.execute("UPDATE student SET marks = 95 WHERE id = 1") # this is the change
conn.commit()

# âœ… Fetch and print the updated data
cur.execute("SELECT * FROM student WHERE id = 1")  # Select the updated record
row = cur.fetchone()  # Fetch one result (since we only expect one record)
print("Updated Record:", row)

conn.close()
```

```python
#DELETE Operation
import sqlite3

conn = sqlite3.connect('students.db')
cur = conn.cursor()

cur.execute("DELETE FROM student WHERE id = 10")
conn.commit()

cur.execute("SELECT * FROM student WHERE id = 10")
row = cur.fetchone() 
print("deleted Record:", row)

# we check that  the row is deleted or not
cur.execute("SELECT * FROM student")
rows = cur.fetchall()  # Fetch all rows
for row in rows:
    print(row)

conn.close()
```

```python
#Handling Errors (Exception Handling)
import sqlite3  #error(important)

conn = sqlite3.connect('school.db')
cur = conn.cursor()

# âŒ Intentionally using wrong table name 'studentss' (extra 's')
cur.execute("SELECT * FROM studentss")

rows = cur.fetchall()
for row in rows:
    print(row)

conn.close()
```

```python
# how to correct it 
import sqlite3

try:
    conn = sqlite3.connect('school.db')     # Try connecting to database
    cur = conn.cursor()

    # âŒ Same wrong query as before
    cur.execute("SELECT * FROM studentss")

    rows = cur.fetchall()
    for row in rows:
        print(row)

except sqlite3.OperationalError as e:        # Handle SQLite operational errors
    print("âŒ Operational Error:", e)        #  e is the variable where we store the error message

finally:
    conn.close()                             # Close database safely
    print("ðŸ”’ Connection closed")
```

```python
#Networking in Python is done using the socket module.
import socket #socket is like a communication endpoint that allows two programs to talk over a network.


#Server  Waits for client request
#Client Sends request to server

#Types of Sockets #TCP Socket( in our syllabus , it is reliable communication)    #UDP socket(not in syllabus)
#Requires two programs:
#1ï¸âƒ£ Server Program
#2ï¸âƒ£ Client Program
```

```python
#TCP Server (order is important)
import socket

server = socket.socket()
server.bind(("localhost", 9999))
server.listen(1)

print("âœ… Server is waiting for connection...")
conn, addr = server.accept()
print("âœ… Connected with:", addr)

data = conn.recv(1024).decode()
print("Client says:", data)

conn.send("Hello Client, Message Received!".encode())
conn.close()
```

```python
#Working with Current Date & Time

#Import the 'datetime' class from the 'datetime' module.
from datetime import datetime

now = datetime.now()#Get the current local date and time
print("Current Date & Time:", now)
print("Date:", now.date())
print("Time:", now.time())
```

```python
#Extracting Date & Time Parts
print("Year:", now.year)
print("Month:", now.month)
print("Day:", now.day)
print("Hour:", now.hour)
print("Minute:", now.minute)
print("Second:", now.second)
```

```python
#Formatting Date & Time
formatted = now.strftime("%d-%m-%Y %H:%M:%S")
print("Formatted Date & Time:", formatted)



# strftime() = "string format time"##converts the datetime object 'now' into a string based on the given format
# %d â†’ Day (2 digits)
# %m â†’ Month (2 digits)
# %Y â†’ Year (4 digits)
# %H â†’ Hour (24-hour format)
# %M â†’ Minutes
# %S â†’ Seconds

```

```python
#Time Delay / Sleep    (Used to pause program execution for some seconds)
import time #Import the time module to use time-related functions

print("Start")
time.sleep(3)  # Wait for 3 seconds
print("3 seconds laterâ€¦")
```

```python
#Program Execution Time (Measure how long a program takes to run)

import time

start = time.time()  ## Record the current time before the loop starts

# Sample code to test
for i in range(5):
    print(i)             #Print numbers 0 to 4
    time.sleep(1)             ## Pause for 1 second between each print

end = time.time()  # Record the current time after the loop ends 
print("Total Execution Time:", end - start, "seconds")
```

```python
#Creating custom date
from datetime import datetime

#Create a datetime object representing the user's birthday
birthday = datetime(2003, 6, 15)
print("Your Birthday:", birthday)
```

```python
#Adding or subtracting time
from datetime import datetime, timedelta

today = datetime.now()            # Get the current date and time
future = today + timedelta(days=10)           # Add 10 days to the current date
past = today - timedelta(days=5)             # Subtract 5 days from the current date

print("10 Days Later:", future)
print("5 Days Ago:", past)
```

```python
#Comparing dates
from datetime import datetime  # Import the datetime class

# Create two datetime objects representing specific dates
d1 = datetime(2025, 1, 1)    # 1st January 2025
d2 = datetime(2024, 12, 31)  # 31st December 2024

# Compare the two dates using comparison operators
if d1 > d2:                  # Checks if d1 is later (after) than d2
    print("d1 is after d2")  # Prints message if condition is True
```
